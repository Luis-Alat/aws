```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('ggplot')
pd.set_option("display.max_columns", None)
```

```python
loan_default_df = pd.read_csv("../data/raw/Loan_default.csv")
loan_default_df.head()
```

```python
loan_default_df.info()
```

```python
# Data Type Definitions

label_col = ["Default"]

num_feat = [
    "Age",
    "Income",
    "LoanAmount",
    "CreditScore",
    "MonthsEmployed",
    "NumCreditLines",
    "InterestRate",
    "LoanTerm",
    "DTIRatio"       
]

cat_feat = [
    "LoanID",
    "Education",
    "EmploymentType",
    "MaritalStatus",
    "HasMortgage",
    "HasDependents",
    "LoanPurpose",
    "HasCoSigner",
]
```

```python
# Drop Feature
(loan_default_df["LoanID"].value_counts() > 1).any()
```

```python
cat_feat.remove("LoanID")
loan_default_df.drop("LoanID", axis=1, inplace=True)
```

```python
# Class imbalance
loan_default_df[label_col].value_counts()
```

# Categorical features

```python
for cf in cat_feat:
    print(loan_default_df[cf].value_counts())
```

```python
default_cat_feat = loan_default_df.query(f"{label_col[0]} == 1")[cat_feat]

row_size_defa_cat_feat = default_cat_feat.shape[0]

for i, ct in enumerate(cat_feat):

    pcg_plot = 100 * (default_cat_feat[ct].value_counts() / row_size_defa_cat_feat)
    pcg_plot = round(pcg_plot, 2)

    plt.figure(figsize=(8, 3))
    
    ax = sns.barplot(pcg_plot, color="steelblue")
    ax.bar_label(ax.containers[0], fmt='%.2f', padding=3) 
    
    plt.ylabel("\nFrequency (porcentages)\n", fontdict={"size": 10})
    plt.show()
```

# Numeric Features

```python
default_num_feat = loan_default_df.query(f"{label_col[0]} == 1")[num_feat]

fig, ax = plt.subplots(len(num_feat), 2, figsize=(10, 50))

for i, nf in enumerate(num_feat):
    
    # Comparing clients distributions with 
    # default vs common distribution (all population) 
    sns.kdeplot(default_num_feat[nf], color="steelblue", ax=ax[i, 0])
    sns.kdeplot(loan_default_df[nf], color="steelblue", ax=ax[i, 1])

plt.show()
```
